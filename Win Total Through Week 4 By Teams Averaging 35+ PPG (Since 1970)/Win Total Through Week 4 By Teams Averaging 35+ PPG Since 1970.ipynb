```python
#try to find number of wins for a team beginning a season averaging at least 35 through 4 games
#represent in a bar chart 
from sportsipy.nfl.teams import Teams
from sportsipy.nfl.teams import Team
from sportsipy.nfl.schedule import Schedule
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
winsarr = []
for i in range(2000, 2022):
    for team in Teams(str(i)):
        j = 0
        pointssum = 0
        wins = 0
        while j < 4:
            pointssum += team.schedule[j].points_scored
            j += 1
        if pointssum >= 140:
            k = 0
            while k < 4:
                if team.schedule[k].result == 'Win':
                    wins += 1 
                k += 1
            winsarr.append(wins)
print(winsarr)
```
